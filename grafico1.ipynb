Número de artículos por universidad (Matplotlib)

```python
#Traer la primera fuente de datos

import pandas as pd
articulos = pd.read_csv("DatosCSV/data.csv", sep=";")
articulos
```

```python
#Traer la segunda fuente de datos

import sqlite3

conn = sqlite3.connect("DatosSQL/students.db")
alumnos = pd.read_sql("select * from personal_data", con=conn)
alumnos
```

```python
datos = pd.merge(articulos, alumnos, on="AlumnoID")
```

```python
datos
```

```python
import matplotlib.pyplot as plt

#Contar artículos por universidad
articulos_por_universidad = datos['NOMBRE UNIVERSIDAD'].value_counts().sort_values(ascending=False)

#Crear gráfico de barras
plt.figure(figsize=(12, 6))  # Ajusta el tamaño del gráfico
articulos_por_universidad.plot(kind='bar', color='skyblue')

#Agregar títulos y etiquetas
plt.title("Número de artículos por Universidad", fontsize=14)
plt.xlabel("Universidad", fontsize=12)
plt.ylabel("Cantidad de artículos", fontsize=12)

#Mostrar el gráfico
plt.show()
```
